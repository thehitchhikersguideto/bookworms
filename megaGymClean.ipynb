### Dealing with None String Values in Equipment Column

```python
# Print the rows that have 'None' string values in the Equipment column
megaGym[megaGym['Equipment'] == 'None']

# print equipment column column values
megaGym['Equipment'].unique()
```

```python
# Change all values that have 'None' in the Equipment column to Body Only if the Type is 'Stretching'
megaGym.loc[(megaGym['Equipment'] == 'None') & (megaGym['Type'] == 'Stretching'), 'Equipment'] = 'Body Only'
equimentNone = megaGym[megaGym['Equipment'] == 'None']
equimentNone



```

```python
# If the title contains a word that is in the Equipment column amd the current value is none, then change the value in the Equipment column to that word, upper and lower case
# Print the changes made
for index, row in equimentNone.iterrows():
    if row['Equipment'] == 'None':
        for word in row['Title'].split():
            if word.lower() in ['bands', 'barbell', 'kettlebells', 'dumbbell', 'cable','machine', 'medicine']:
                equimentNone.loc[index, 'Equipment'] = word.title()
                print(row['Title'], 'changed to', word.title())
                break

# Merge the changes made to the original data frame
megaGym.update(equimentNone)

# If the Type is 'Pylometrics' and equipment is 'None' then change the value of equipment to 'Body Only'
megaGym.loc[(megaGym['Equipment'] == 'None') & (megaGym['Type'] == 'Plyometrics'), 'Equipment'] = 'Body Only'
# And the same for 'Strength' 
megaGym.loc[(megaGym['Equipment'] == 'None') & (megaGym['Type'] == 'Strength'), 'Equipment'] = 'Body Only'

# Should be empty
megaGym[megaGym['Equipment'] == 'None']
```

### Balancing Levels

```python
# The levels column is greatly imbalanced, we will redistribute the values to make it more balanced
megaGym['Level'].value_counts()
```

```python
# We want more of the 'Beginner' and 'Expert' values, so we will replace some of the 'Intermediate' values with 'Beginner' and 'Expert'
# We will replace 300 'Intermediate' values with 'Beginner' and 300 'Intermediate' values with 'Expert'
# Realistically we wouldnt want to do this randomly as it would not be representative of the data, but for the sake of this exercise we will do it randomly

# Get the indices of the rows that have 'Intermediate' values
intermediateIndices = megaGym[megaGym['Level'] == 'Intermediate'].index
# Randomly select 2 sets of 300 indices from the intermediate indices, make sure that the indices are not repeated
beginnerIndices = np.random.choice(intermediateIndices, 300, replace=False)
expertIndices = np.random.choice(intermediateIndices, 300, replace=False)
# Change the values of the indices to 'Beginner' and 'Expert'
megaGym.loc[beginnerIndices, 'Level'] = 'Beginner'
megaGym.loc[expertIndices, 'Level'] = 'Expert'

# Check the values
megaGym['Level'].value_counts()
```
